```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.cluster import KMeans
```

# load dataset 

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df["target"]=iris.target
```

```python
df
```

```python
# Assign target names to a new column "Flowers"
df["Flowers"] = iris.target_names[iris.target]
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
obj=LabelEncoder()
```

```python
df["Flowers"]=obj.fit_transform(df["Flowers"])
df
```

# Cluster Visualization 

```python
 
# Plotting scatter plot
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], marker="*", color="red", label="Sepal")
plt.scatter(df['petal length (cm)'], df['petal width (cm)'], marker="o", color="blue", label="Petal")

# Adding labels and legend
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('Scatter plot of Sepal and Petal Dimensions')
plt.legend()

# Display the plot
plt.show()
```

```python
cluster=KMeans(n_clusters=3)
```

```python
df["Clusters"]=cluster.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
```

```python
df
```

```python
df1=df[df.target==0]
df2=df[df.target==1]
df3=df[df.target==2]
```

```python
 
# Plotting scatter plot
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], marker="*", color="red", label="Sepal")
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], marker="o", color="blue", label="Petal")
plt.scatter(df3['petal length (cm)'], df3['petal width (cm)'], marker="o", color="black", label="Petal")

# Adding labels and legend
plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('Scatter plot of Sepal and Petal Dimensions')
plt.legend()

# Display the plot
plt.show()
```

```python
df
```

# Elbow Curve

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['sepal length (cm)','sepal width (cm)']])
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
